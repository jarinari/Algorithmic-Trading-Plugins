```python
import random
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
n = 200 # Cambia esto al número de elementos que quieras seleccionar
lista = ["cara"] * 1000 + ["cruz"] * 1000
```

```python
ave_abs = []
ave_rel = []
for n in tqdm(range(1,1000)):
  absolutes = []
  relatives = []
  for x in range(1000):
    elementos_aleatorios = random.sample(lista, n)
    cara_count = elementos_aleatorios.count("cara")
    cruz_count = elementos_aleatorios.count("cruz")
    dif_abs = abs(cara_count - cruz_count)
    absolutes.append(dif_abs)
    dif_rel = abs((abs(cara_count-(n/2))/(n/2)))
    relatives.append(dif_rel)
  ave_abs.append(np.mean(absolutes))
  ave_rel.append(np.mean(relatives))
```

```python
fig, ax1 = plt.subplots(figsize=(10,5))
ax1.plot(ave_abs, 'g-')
ax1.set_ylabel('Diferencia Absoluta', color='g')
ax1.set_xlabel('Número de tiradas')
ax2 = ax1.twinx()
ax2.plot(ave_rel, 'b-')
ax2.set_ylabel('Diferencia Relativa', color='b')
plt.title('DIFERENCIA ABSOLUTA vs DIFERENCIA RELATIVA EN FUNCIÓN DE LAS TIRADAS')
plt.show()
```
